```python
import csv
import os

class CSVProcessor:
    def __init__(self, fdir, fname):
        self.fdir = fdir
        self.fname = fname
        
    def process(self):
        line = 0
        result = []
        with open(os.path.join(self.fdir, self.fname)) as raw:
            parsed = csv.reader(raw, delimiter=',')
            for row in parsed:
                if line > 0:
                    result.append((row[0], bool(row[1])))
                line += 1
        return result
```

```python
proc = CSVProcessor('data/', 'trainLabels.csv')
proc.process()
```
