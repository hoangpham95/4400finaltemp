```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import validation_curve
from sklearn.linear_model import Ridge

import pandas as pd
import numpy as np
```

```python
df_500 = pd.read_csv(filepath_or_buffer="data/train500.csv", index_col=0)
test_500 = pd.read_csv(filepath_or_buffer="data/test500.csv", index_col=0)

dropped = ['file', 'label']

X = df_500.drop(columns=dropped)
y = df_500['label']
```

```python
rdc = RandomForestClassifier(n_estimators=600, max_depth=10)
rdc.fit(X, y)
```

```python
rdc.score(X, y)
```

```python
rdc.score(test_500.drop(columns=dropped), test_500['label'])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
```

```python
accuracy_score(test_500['label'], rdc.predict(test_500.drop(columns=dropped)))
```

```python
log_loss(test_500['label'], rdc.predict_proba(test_500.drop(columns=dropped)))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dcf = DecisionTreeClassifier(max_depth=8)
dcf.fit(X, y)
```

```python
X_test, y_test = test_500.drop(columns=dropped), test_500['label']
```

```python
dcf.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

dcf = DecisionTreeClassifier(max_depth=8)
cross_val_score(dcf, X, y, cv=5)
```

```python
import matplotlib.pyplot as plt

depths = [8, 10, 12, 16, 20]
colors = ['r', 'g', 'b', 'c', 'm']

for i in range(len(depths)):
    d = depths[i]
    dcf = DecisionTreeClassifier(max_depth=d)
    x = range(10)
    score = cross_val_score(dcf, X, y, cv=len(x))
    
    plt.plot(x, score, colors[i], label='Depth ' + str(d))
    plt.legend(loc='upper left')
```

```python
from sklearn.model_selection import GridSearchCV
from time import time
```

```python
param_grid = {
    "max_depth": [None, 8, 10, 12, 16, 20],
    "n_estimators": [100, 200, 400, 600, 800],
    "criterion": ["gini", "entropy"]
}

clf = RandomForestClassifier()
grid= GridSearchCV(clf, param_grid=param_grid, cv=10)
start = time()
grid.fit(X, y)
print("GridSearchCV took %.2f seconds"
      % (time() - start))
```

```python
grid.best_estimator_
```

```python
rdc = grid.best_estimator_
print(rdc.score(X, y))
print(rdc.score(X_test, y_test))
```

```python
n_estimators = [100, 200, 500, 600, 800]
scores = []
preds = []
for est in n_estimators:
    ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=2), n_estimators=est, learning_rate=0.3)
    ada.fit(X, y)
    scores.append(ada.score(X, y))
    preds.append(ada.score(X_test, y_test))
    
plt.plot(n_estimators, scores, 'r', label='AdaBoost Train Acc')
plt.plot(n_estimators, preds, 'b', label='AdaBoost Test Acc')
plt.legend(loc='lower right')
```
